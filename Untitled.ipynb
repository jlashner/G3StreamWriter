```python
from spt3g import core
import matplotlib.pyplot as plt
import numpy as np
from collections import defaultdict
```

```python


for i in range(10):
    file = core.G3File("phases.g3")
    for frame in file:
        ts = frame["Phases"][str(i)]
        timestamps = np.linspace(ts.start.time, ts.stop.time, len(ts))
        plt.plot(timestamps, ts)
plt.show()
```
